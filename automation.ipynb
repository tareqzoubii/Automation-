```python
# Read Data:
import pandas as pd
df=pd.read_csv("./data.csv")
df
```

```python
# Split Data:
df_1 , df_2 = df[:1], df[1:]
meeting_duration= int(df_1["Duration (Minutes)"][0])
meeting_duration
```

```python
# Prepare Data
df_2.columns=df_2.iloc[0]
df_2.drop(df_2.index[0],inplace=True)
df_2
```

```python
# Student ID:
df_2["Student ID"]= df_2["User Email"].str.extract(pat='(^[0-9]*)')
df_2["Student ID"]
```

```python
# String to Int!
df_2["Total"]= df_2["Duration (Minutes)"].astype("int")
df_2["Total"]
```

```python
# Total Time with ID:
df_2= df_2.groupby("Student ID",as_index=False).sum(numeric_only=True)
df_2
```

```python
# Status
df_2["Status"] = ['Present' if x >= meeting_duration-10 else "Absent" if x <  meeting_duration-10 and x >= meeting_duration-20 else "Late" for x in df_2["Total"]]
df_2
```
